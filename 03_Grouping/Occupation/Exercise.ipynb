### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user).

### Step 3. Assign it to a variable called users.

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
users = pd.read_csv(url, sep = '|')
```

```python
users.head(2)
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby(by = 'occupation').agg({'age':'mean'})
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
df = users.groupby(by = ['occupation', 'gender'], as_index = True).agg({'gender':'count'}).rename(columns={'gender':'count'})
df
```

```python
df = users.groupby(by = ['occupation', 'gender'], as_index = True).agg({'gender':'count'}).rename(columns={'gender':'count'})
left_df = df.xs('M', level = 'gender').reset_index()
right_df = df.xs('F', level='gender').reset_index()
left_df.merge(right_df, on='occupation', how='outer').rename(columns = {'count_x':'M', 'count_y':'F'}).fillna(0).assign(total = lambda x: x['F'] + x['M']).assign(m_ratio = lambda x: round(x['M']/x['total'], 2)).sort_values(by = 'm_ratio', ascending = False).iloc[:, [0, 4]]
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby('occupation').agg({'age':['min', 'max']})
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(by = ['occupation', 'gender']).agg({'age':'mean'})
```

### Step 8.  For each occupation present the percentage of women and men

```python
left_df.merge(right_df, on='occupation', how='outer').rename(columns = {'count_x':'M', 'count_y':'F'}).fillna(0).assign(total = lambda x: x['F'] + x['M']).assign(m_ratio = lambda x: round(x['M']/x['total'], 2)).assign(f_ratio = lambda x: round(x['F']/x['total'],2)).sort_values(by = 'm_ratio', ascending = False).iloc[:, [0, 4, 5]]
```

```python
gender_ocup = users.groupby(['occupation', 'gender']).agg({'gender': 'count'})

# create a DataFrame and apply count for each occupation
occup_count = users.groupby(['occupation']).agg('count')

# divide the gender_ocup per the occup_count and multiply per 100
occup_gender = gender_ocup.div(occup_count, level = "occupation") * 100

# present all rows from the 'gender column'
occup_gender.loc[: , 'gender']
```
